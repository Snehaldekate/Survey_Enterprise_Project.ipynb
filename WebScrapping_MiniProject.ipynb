```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
import os.path
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```python
Product_Name = []
Price = []
StarRating = []
Details = []
```

```python
for i in range(1,21):
  m_url = requests.get(f'https://www.flipkart.com/search?q=mobile+&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}').text
  soup = bs(m_url,'lxml')

for mobilephone in soup.find_all('a',class_="_1fQZEK"):
  mob_name = mobilephone.find('div',class_="_4rR01T")
  if mob_name != None:
    Product_Name.append(mob_name.text)
  else:
    Product_Name.append("none")

  mob_price = mobilephone.find('div',class_="_30jeq3 _1_WHN1")
  if mob_price != None:
    Price.append(mob_price.text)
  else:
    Price.append("None")

  mobile_rating = mobilephone.find('div',class_="_3LWZlK")
  if mobile_rating != None:
    StarRating.append(mobile_rating.text)
  else:
    StarRating.append("None")

  complete_details = ""
  for details in mobilephone.find_all("li",class_="rgWa7D"):
    complete_details += details.text+","
  Details.append(complete_details)
```

```python
df = pd.DataFrame({'MobileName':Product_Name,
   'StarRating':StarRating,
   'Price':Price,
   'Details':Details
   })
df
```

```python
file_name = 'Mobilephone.xlsx'
# saving the excel
df.to_excel(file_name)
print('DataFrame is written to Excel File successfully.')
```
